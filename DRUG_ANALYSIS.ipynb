```python
!pip install PyTDC

import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Conv1D, GlobalMaxPooling1D, Dense, LayerNormalization, MultiHeadAttention, Dropout, Multiply, Add
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tdc.multi_pred import DTI

df = pd.read_csv('davis_clean.csv')

df['Drug'] = df['Drug'].fillna('')
df['Target'] = df['Target'].fillna('')

def neighbor_avg_impute(series):
    series = series.copy()
    global_mean = series.mean()
    for i in range(len(series)):
        if pd.isna(series[i]):
            prev_val = series[i-1] if i > 0 else np.nan
            next_val = series[i+1] if i < len(series)-1 else np.nan
            neighbors = [v for v in [prev_val, next_val] if not pd.isna(v)]
            if neighbors:
                series[i] = np.mean(neighbors)
            else:
                series[i] = global_mean
    return series

df['Y'] = neighbor_avg_impute(df['Y'])

print("Missing values after imputation:")
print(df.isna().sum())
MAX_LEN_SMILES = 100
MAX_LEN_PROTEIN = 1000
VOCAB_SIZE = 40


smiles_tokenizer = Tokenizer(char_level=True, num_words=VOCAB_SIZE, oov_token='O')
smiles_tokenizer.fit_on_texts(df['Drug'])
X_smiles = pad_sequences(smiles_tokenizer.texts_to_sequences(df['Drug']), maxlen=MAX_LEN_SMILES)


protein_tokenizer = Tokenizer(char_level=True, num_words=VOCAB_SIZE, oov_token='O')
protein_tokenizer.fit_on_texts(df['Target'])
X_protein = pad_sequences(protein_tokenizer.texts_to_sequences(df['Target']), maxlen=MAX_LEN_PROTEIN)


Y = np.log1p(df['Y'].values)


X_train_sm, X_test_sm, X_train_prot, X_test_prot, Y_train, Y_test = train_test_split(
    X_smiles, X_protein, Y, test_size=0.2, random_state=42
)

input_sm = Input(shape=(MAX_LEN_SMILES,))
embed_sm = Embedding(VOCAB_SIZE, 64)(input_sm)
conv_sm = Conv1D(64, 5, activation='relu')(embed_sm)
pool_sm = GlobalMaxPooling1D()(conv_sm)

input_prot = Input(shape=(MAX_LEN_PROTEIN,))
embed_prot = Embedding(VOCAB_SIZE, 64)(input_prot)
attn = MultiHeadAttention(num_heads=2, key_dim=64)(embed_prot, embed_prot)
attn = LayerNormalization()(attn)
attn_pool = GlobalMaxPooling1D()(attn)

sm_weight = Dense(1, activation='sigmoid')(pool_sm)
prot_weight = Dense(1, activation='sigmoid')(attn_pool)
weighted_sm = Multiply()([pool_sm, sm_weight])
weighted_prot = Multiply()([attn_pool, prot_weight])
fused = Add()([weighted_sm, weighted_prot])

dense = Dense(128, activation='relu')(fused)
drop = Dropout(0.2)(dense)
output = Dense(1)(drop)

model = Model(inputs=[input_sm, input_prot], outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()


history = model.fit([X_train_sm, X_train_prot], Y_train,
                    validation_data=([X_test_sm, X_test_prot], Y_test),
                    epochs=5, batch_size=32)


plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss (Davis)")
plt.legend()
plt.grid()
plt.show()

Y_pred = model.predict([X_test_sm, X_test_prot])
Y_test_orig = np.expm1(Y_test)
Y_pred_orig = np.expm1(Y_pred.flatten())

plt.figure(figsize=(6,6))
plt.scatter(Y_test_orig, Y_pred_orig, alpha=0.5, color='purple')
plt.plot([min(Y_test_orig), max(Y_test_orig)],
         [min(Y_test_orig), max(Y_test_orig)],
         'k--')
plt.xlabel("Actual Affinity (Davis)")
plt.ylabel("Predicted Affinity")
plt.title("Prediction vs Ground Truth (Davis)")
plt.grid()
plt.show()

kiba = DTI(name='KIBA').get_data()

kiba['Drug'] = kiba['Drug'].fillna('')
kiba['Target'] = kiba['Target'].fillna('')
kiba['Y'] = neighbor_avg_impute(kiba['Y'])

X_kiba_sm = pad_sequences(smiles_tokenizer.texts_to_sequences(kiba['Drug']), maxlen=MAX_LEN_SMILES)
X_kiba_prot = pad_sequences(protein_tokenizer.texts_to_sequences(kiba['Target']), maxlen=MAX_LEN_PROTEIN)
Y_kiba = np.log1p(kiba['Y'].values)

Y_kiba_pred = model.predict([X_kiba_sm, X_kiba_prot])
Y_kiba_true = np.expm1(Y_kiba)
Y_kiba_pred_orig = np.expm1(Y_kiba_pred.flatten())

plt.figure(figsize=(6,6))
plt.scatter(Y_kiba_true, Y_kiba_pred_orig, alpha=0.5, color='teal')
plt.plot([min(Y_kiba_true), max(Y_kiba_true)],
         [min(Y_kiba_true), max(Y_kiba_true)],
         'k--')
plt.xlabel("Actual Affinity (KIBA)")
plt.ylabel("Predicted Affinity (from Davis model)")
plt.title("Cross-Dataset Evaluation on KIBA")
plt.grid()
plt.show()

mae_kiba = mean_absolute_error(Y_kiba_true, Y_kiba_pred_orig)
print("Cross-Dataset MAE on KIBA:", mae_kiba)
```
